```python
import dotfiles as dfs
from PIL import Image, ImageDraw
import requests
from bs4 import BeautifulSoup
```

# Image

```python
def crop_circle(image_path, save_path):
    diameter = 1000
    image = Image.open(image_path)
    width, height = image.size
    radius = diameter // 2
    center_x = width // 2
    center_y = height // 2
    left = center_x - radius
    top = center_y - radius
    right = center_x + radius
    bottom = center_y + radius
    cropped_image = Image.new("RGBA", (diameter, diameter), (0, 0, 0, 0))
    circle_mask = Image.new("L", (diameter, diameter), 0)
    draw = ImageDraw.Draw(circle_mask)
    draw.ellipse((0, 0, diameter, diameter), fill=255)
    cropped_circle = image.crop((left, top, right, bottom)).resize((diameter, diameter))
    cropped_image.paste(cropped_circle, (0, 0), mask=circle_mask)
    # resized_image = cropped_image.resize((120, 120), resample=Image.LANCZOS)
    cropped_image.save(save_path, "PNG", quality=1000)
```

```python
url = "https://mafia1986.com/game/scenario/12/"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
image_tags = soup.find_all("img", class_="object-contain")
image_data = [{"url": img["src"], "alt": img.get("alt", "")} for img in image_tags]
```

```python
for data in image_data:
    url = f"https://mafia1986.com{data['url']}"
    name = data["alt"].replace("*", "").replace(" ", "")
    dfs.utils.download(url, f"/Users/farid/Desktop/shab/{name}.jpg")
    crop_circle(
        f"/Users/farid/Desktop/shab/{name}.jpg",
        f"/Users/farid/Desktop/shab/{name}.png",
    )
    break
```

```python
url = "https://mafia1986.com/game/scenario/12/"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
card_bodies = soup.find_all("div", class_="card-body")


def reformat(text):
    text = text.replace(" :", ":")
    text = text.replace("   :", ":")
    text = text.replace("( ", "(")
    text = text.replace(" )", ")")
    text = text.replace(" .", ".")
    # text = text.replace("می ", "می‌")
    # text = text.replace(" ای", "‌ای")
    text = text.replace("میشود", "می‌شود")
    text = text.replace(" !", "!")
    return text


results = {}
for card_body in card_bodies:
    combined_text = ""
    name = ""
    h2_tag = card_body.find("h2")
    if h2_tag:
        name = h2_tag.text.replace("*", "").replace(" ", "").replace("آ", "ا")
    div_w_80vw = card_body.find("div", class_="w-[80vw]")
    if div_w_80vw:
        h3_tag = div_w_80vw.find("h3")
        if h3_tag:
            combined_text += f"{reformat(h3_tag.text)}<br/>"
        p_tags = div_w_80vw.find_all("p")
        div_tags = div_w_80vw.find_all("div")
        h5_tags = div_w_80vw.find_all("h5")
        if p_tags:
            combined_text += "<br/>".join([reformat(p.text) for p in p_tags])
            results[name] = combined_text
        elif div_tags:
            combined_text += "<br/>".join([reformat(p.text) for p in div_tags])
            results[name] = combined_text
        elif h5_tags:
            combined_text += "<br/>".join([reformat(p.text) for p in h5_tags])
            results[name] = combined_text
        else:
            results[name] = None
```
